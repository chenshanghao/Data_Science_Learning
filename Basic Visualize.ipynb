```python
import numpy as np
import pandas as pd
from pandas import DataFrame, Series
import pymongo
from pymongo import MongoClient
import pprint
import re

%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
```

#### 1. Connect the Mongodb

```python
client = MongoClient()
db_ctTrends = client.ctTrends
db_tlMetadata = client.tlMetadata
```

#### 2. Get the trend log data from Mongodb

```python
# Get the date
regx = re.compile("^3352\.750000\.TL16-201[5678] *", re.IGNORECASE)  #\.*-2018-06-*\
data_example = db_ctTrends.Latest.find({'_id': regx}).limit(1000)
data_example_list = list(data_example)
pprint.pprint(data_example_list[0])
```

#### 3. Transfer the numerical value data to pandas Series(Need to optimize?)

```python
from datetime import datetime

# Convert date string to datetime
def datetime_convert(a, b):
    return datetime.strptime(a + b, '%Y-%m-%d%H:%M:%S')

pat = re.compile(r'\d{2}:\d{2}:\d{2}')
data_example_pandas = (pd.DataFrame([[datetime_convert(r['dt'],k), v] for r in data_example_list for k, v in r.items() if pat.match(k)], 
                                    columns=['Time-Stamp', 'Value']).set_index('Time-Stamp')['Value'])

# print the first ten lines of data
print(data_example_pandas[0:10])
```

#### 4. Get the meta data from Mongodb

```python
regx = re.compile("^3352\.750000\.TL16", re.IGNORECASE)  #\.*-2018-06-*
data_example_meta = pd.DataFrame(list(db_tlMetadata.Latest.find({'_id': regx})))
data_example_meta
```

#### 5. Plotting the Figure

```python
fig, ax = plt.subplots(figsize=(20, 10))
ax.plot(data_example_pandas.index.values, data_example_pandas.values, label=str(data_example_meta['_id'].values[0]))
leg = ax.legend();
```
