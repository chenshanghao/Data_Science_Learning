```python
import matplotlib.pyplot as plt
plt.style.use('classic')
import numpy as np

%matplotlib inline

x = np.random.randn(1000)
plt.hist(x);
```

# Plot Customization by Hand

```python
# use a gray background
ax = plt.axes(facecolor='#E6E6E6')
ax.set_axisbelow(True)

# draw solid white grid lines
plt.grid(color='w', linestyle='solid')

# hide axis spines
for spine in ax.spines.values():
    spine.set_visible(False)    # spine cannot see
    
# hide top and right ticks
ax.xaxis.tick_bottom()
ax.yaxis.tick_left()

# lighten ticks and labels
ax.tick_params(colors='gray', direction='out')
for tick in ax.get_xticklabels():
    tick.set_color('gray')
for tick in ax.get_yticklabels():
    tick.set_color('gray')
    
# control face and edge color of histogram
ax.hist(x, edgecolor='#E6E6E6', color='#EE6666');
```

# Changing the Defaults: rcParams


```python
# This configuration can be adjusted at any time using the plt.rc convenience routine.
IPython_default = plt.rcParams.copy()

# Now we can use the plt.rc function to change some of these settings:

from matplotlib import cycler
colors = cycler('color',
                ['#EE6666', '#3388BB', '#9988DD',
                 '#EECC55', '#88BB44', '#FFBBBB'])
plt.rc('axes', facecolor='#E6E6E6', edgecolor='none',
       axisbelow=True, grid=True, prop_cycle=colors)
plt.rc('grid', color='w', linestyle='solid')
plt.rc('xtick', direction='out', color='blue')
plt.rc('ytick', direction='out', color='green')
plt.rc('patch', edgecolor='#E6E6E6')
plt.rc('lines', linewidth=2)

# hidden
a = plt.axes()
a.xaxis.tick_bottom()
a.yaxis.tick_left()


# create a plot and see our settings in action:
plt.hist(x);
```

```python
# Simple line plots look like with these rc parameters
for i in range(4):
    plt.plot(np.random.rand(10))
```

# Default Style

```python
def hist_and_lines():
    np.random.seed(0)
    fig, ax = plt.subplots(1, 2, figsize=(11, 4))
    ax[0].hist(np.random.randn(1000))
    for i in range(3):
        ax[1].plot(np.random.rand(10))
    ax[1].legend(['a', 'b', 'c'], loc='lower left')
```

```python
# reset rcParams
plt.rcParams.update(IPython_default)
```

```python
hist_and_lines()
```

```python
with plt.style.context('fivethirtyeight'):
    hist_and_lines()
```
